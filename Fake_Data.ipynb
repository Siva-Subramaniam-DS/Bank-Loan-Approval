```python
from faker import Faker
import pandas as pd
import random

fake = Faker('en_IN')
num_samples = 1000

data = []
for _ in range(num_samples):
    customer_id = fake.uuid4()
    name = fake.name()
    age = random.randint(21, 65)
    gender = random.choice(["Male", "Female"])
    phone_number = fake.phone_number()
    email = fake.email()
    address = fake.address().replace("\n", ", ")
    city = fake.city()
    state = fake.state()
    pincode = fake.postcode()
    bank_balance = random.randint(10000, 2000000)
    cibil_score = random.randint(300, 900)
    existing_loans = random.choice(["Yes", "No"])
    loan_amount_requested = random.randint(50000, 1000000)
    loan_tenure_months = random.choice([12, 24, 36, 48, 60])
    monthly_income = random.randint(20000, 300000)
    income_source = random.choice(["Salary", "Business", "Rental", "Freelance"])
    employment_type = random.choice(["Salaried", "Self-employed"])
    other_loans = random.randint(0, 500000) if existing_loans == "Yes" else 0
    emi_existing_loans = random.randint(0, 50000) if existing_loans == "Yes" else 0
    loan_status = random.choice(["Approved", "Rejected"])

    data.append([customer_id, name, age, gender, phone_number, email, address, city, state, pincode, 
                 bank_balance, cibil_score, existing_loans, loan_amount_requested, loan_tenure_months,
                 monthly_income, income_source, employment_type, other_loans, emi_existing_loans, loan_status])

columns = ["Customer_ID", "Name", "Age", "Gender", "Phone_Number", "Email", "Address", "City", "State", "Pincode", 
           "Bank_Balance", "CIBIL_Score", "Existing_Loans", "Loan_Amount_Requested", "Loan_Tenure_Months",
           "Monthly_Income", "Income_Source", "Employment_Type", "Other_Loans", "EMI_Existing_Loans", "Loan_Status"]

df = pd.DataFrame(data, columns=columns)
df.to_csv("Loan_Approval_Indian_Customers.csv", index=False)
print("Excel file saved as Loan_Approval_Indian_Customers.csv")
```
